```python
# importing the pandas library

import pandas as pd
```

```python
# assigning the csv file web link to a variable "url"

url='https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv'

# using the pandas library to read the csv file and assigning to the variable "fuel_data"

fuel_data = pd.read_csv(url, error_bad_lines = False)
```

```python
# loading the "fuel_data" DataFrame

fuel_data
```

```python
# Direct way to know the "fuel type code" with the lowest average "fuel cost per unit burned".
# in this case, gas with an avearge "fuel cost per unit burned" of 13.659 is the lowest
# when comapred to other fuel type code.

fuel_data.groupby('fuel_type_code_pudl')['fuel_cost_per_unit_burned'].mean()
```

```python
# Alternative way to know the "fuel type code" with the lowest average "fuel cost per unit burned".
# in this case, gas with an avearge "fuel cost per unit burned" of 13.659 is the lowest.

fuel_data.groupby('fuel_type_code_pudl').mean()
```

```python
# getting the standard deviation and 75th percentile 
# of "fuel_mmbtu_per_unit" from the "fuel_data" DataFrame.
# in this case are 10.60 and 17.01 respectively to 2 decimal place.

fuel_data.describe()
```

```python
# Alternative/direct way to get the standard deviation and 75th percentile
# of "fuel_mmbtu_per_unit" from the "fuel_data" DataFrame.
# in this case are 10.60 and 17.01 respectively to 2 decimal place.

fuel_data['fuel_mmbtu_per_unit'].describe()
```

```python
# Direct way of getting the kurtosis for the "fuel quantity burned"

fuel_data['fuel_qty_burned'].kurt()
```

```python
# Alternative way of getting the kurtosis value for the "fuel quantity burned"
# in this case the kurtosis value is 651.369

fuel_data.kurt()
```

```python
# Direct way of getting the skewness for the "fuel quantity burned"

fuel_data['fuel_qty_burned'].skew()
```

```python
# Alternative way of getting the skewness for the "fuel quantity burned"
# in this case the skewness value is 15.851

fuel_data.skew()
```

```python
# to know the feature(s) that has missing values,
# the total number of missing value and percentage of the missing rows
# as a factor of the total number of rows in three decimal places

fuel_data.isnull().sum()

# NB: answer is in this format (Feature: xxx, Total: xxx, Percent: xxx)
# Feature: fuel_unit, Total: 180, Percent: 0.006

# NB: from the result, the feature with missing values falls under "Categorical" category
# "mode imputation" is the missing value imputation technique that should be used.
```

```python
# Direct way to know the year with the highest average "fuel cost per unit delivered"
# in this case, it is 1997 with a value of 11140.197 to 3 d.p

fuel_data.groupby('report_year')['fuel_cost_per_unit_delivered'].mean()
```

```python
# Alternative way to know the year with the highest average "fuel cost per unit delivered"
# in this case, it is 1997 with a value of 11140.197 to 3 d.p

fuel_data.groupby('report_year').mean()
```

```python
# to find the correlation between the "fuel_data" features

fuel_data.corr()
```
